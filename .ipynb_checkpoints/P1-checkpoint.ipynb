```python
# Global parameters

# Gaussian smoothing
kernel_size = 3

# Canny edge detector - now using auto selection of params below
#low_threshold = 100
#high_threshold = 200

# Region of interest vertices
# We want a trapezoid shape, with bottom edge at the bottom of the image
#mask_region_bottom_width = 0.9   # width of bottom edge of trapezoid, expresses as % of image width
#mask_region_top_width = 0.1
#mask_region_height = 0.38

mask_region_bottom_width = 0.9   # width of bottom edge of trapezoid, expresses as % of image width
mask_region_top_width = 0.1
mask_region_height = 0.38
mask_region_bottom = 0.1

# Hough Transform
 = 0.33             # distance resolution in pixels of the Hough grid
 = np.pi/180        # angular resolution in radians of the Hough grid
threshold = 10       # minimum number of votes (intersections in Hough grid cell) - going above 15 makes the dotted line section jittery
min_line_length = 10 # minimum number of pixels making up a line
max_line_gap = 25    # maximum gap in pixels between connectable line segments

    
def auto_canny(image, =0.33):
    """
    Calculate Canny edges using auto parameter selection
    http://www.pyimagesearch.com/2015/04/06/zero-parameter-automatic-canny-edge-detection-with-python-and-opencv/
    Seems to be marginally better than my choice of 100 to 200. Values close to these make very little
    difference in the output. The Hough params are much more sensitive.
    """
    # Compute the media of the single channel pixel intensities
    v = np.median(image)
    
    # Apply automatic canny edge detection using the computed median
    lower = int(max(0, (1.0 - ) * v))
    upper = int(min(255, (1.0 + ) * v))
    edged = cv2.Canny(image, lower, upper)
    
    return edged
```

Below are some helper functions to help get you started. They should look familiar from the lesson!

```python
import math

def grayscale(img):
    """Applies the Grayscale transform
    This will return an image with only one color channel
    but NOTE: to see the returned image as grayscale
    you should call plt.imshow(gray, cmap='gray')"""
    return cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    
def canny(img, low_threshold, high_threshold):
    """Applies the Canny transform"""
    return cv2.Canny(img, low_threshold, high_threshold)

def gaussian_blur(img, kernel_size):
    """Applies a Gaussian Noise kernel"""
    return cv2.GaussianBlur(img, (kernel_size, kernel_size), 0)

def region_of_interest(img, vertices):
    """
    Applies an image mask.
    
    Only keeps the region of the image defined by the polygon
    formed from `vertices`. The rest of the image is set to black.
    """
    #defining a blank mask to start with
    mask = np.zeros_like(img)   
    
    #defining a 3 channel or 1 channel color to fill the mask with depending on the input image
    if len(img.shape) > 2:
        channel_count = img.shape[2]  # i.e. 3 or 4 depending on your image
        ignore_mask_color = (255,) * channel_count
    else:
        ignore_mask_color = 255
        
    #filling pixels inside the polygon defined by "vertices" with the fill color    
    cv2.fillPoly(mask, vertices, ignore_mask_color)
    
    #returning the image only where mask pixels are nonzero
    masked_image = cv2.bitwise_and(img, mask)
    return masked_image


def fit_and_draw_line(img, lines, color, thickness):
    '''
    Helper function to fit line to a set of left/right lines
    and draw the calculated line on the given image.
    '''
    
    # Get the two y points of the straight lines we will draw - just from
    # bottom of image up to the mask region height.
    draw_line_y1 = img.shape[0]
    draw_line_y2 = img.shape[0] * (1 - mask_region_height)
    
    x_points = []
    y_points = []
    
    for line in lines:
        x1, y1, x2, y2 = line[0]
        x_points.append(x1)
        x_points.append(x2)
        y_points.append(y1)
        y_points.append(y2)
    
    if len(x_points) > 0:
        m, b = np.polyfit(x_points, y_points, 1)
        
        # Calculate x point by re-arranging: y = mx + b
        draw_line_x1 = (draw_line_y1 - b) / m
        draw_line_x2 = (draw_line_y2 - b) / m
        
        point1_as_int = (int(draw_line_x1), int(draw_line_y1))
        point2_as_int = (int(draw_line_x2), int(draw_line_y2))
        
        cv2.line(img, point1_as_int, point2_as_int, color, thickness)
        
    
def draw_lines(img, lines, color=[255, 0, 0], thickness=10):
    """
    Draw a pair of straight lines as a best fit of the line vectors provided in 'lines':
    e.g.
    TYPE:  <class 'numpy.ndarray'> (13, 1, 4)
    [[[182 517 390 377]]
     [[179 538 391 379]]
     [[181 518 391 377]]
     [[179 537 341 415]]
     [[525 342 560 362]]]
    
    ...onto the image img.
    
    NOTE: I have no ida why the array is 3 dimensions - I think it is because numpy
    keeps the third (colour) dimension even though we have grayscaled the image!
    
    Defaults to a red line thickness 10.
    
    Split lines into right_lines and left_lines, representing the right and left lane lines
    We can use the gradient of each line to determine if its from the right or left lane - 
    additionally we check that the left lane points are actually from the left-side of the 
    image and likewise for the right lane points. We must do this otherwise we'd associate
    the line with the wrong lane on turns.
    Additionally - we are filtering the gradients which is necessary with the challenge video
    due to its shading of sections of the road as well as horizontal road colourings. It goes
    bezerk without this...
    """
    # Original code which just drew the lines as passed in
    #for line in lines:
    #    for x1,y1,x2,y2 in line:
    #        cv2.line(img, (x1, y1), (x2, y2), color, thickness)
    
    image_center_x = img.shape[1] / 2   # x coord of center of image
    
    right_lines = []
    left_lines = []
    
    for line in lines:
        x1, y1, x2, y2 = line[0]          # the lines are wrapped in another dimension!
        
        try:
            line_gradient = (y2 - y1) / (x2 - x1)
            if line_gradient > 0 and x1 > image_center_x and x2 > image_center_x:
                if 0.45 <= line_gradient <= 0.75:
                    right_lines.append(line)
            elif line_gradient < 0 and x1 < image_center_x and x2 < image_center_x:
                if -0.85 <= line_gradient <= -0.6:
                    left_lines.append(line)
            
        except ZeroDivisionError:
            
            pass  #ignore for now - should we just add a high +ve or -ve high value???

    
    # Fit a straight line of best fit to the points for right and left lane lines
    fit_and_draw_line(img, right_lines, color, thickness)
    fit_and_draw_line(img, left_lines, color, thickness)


def hough_lines(img, rho, theta, threshold, min_line_len, max_line_gap):
    """
    `img` should be the output of a Canny transform.
        
    Returns an image with hough lines drawn.
    """
    lines = cv2.HoughLinesP(img, rho, theta, threshold, np.array([]), minLineLength=min_line_len, maxLineGap=max_line_gap)
    line_img = np.zeros((*img.shape, 3), dtype=np.uint8)
    draw_lines(line_img, lines)
    return line_img

# Python 3 has support for cool math symbols.

def weighted_img(img, initial_img, =0.8, =1., =0.):
    """
    `img` is the output of the hough_lines(), An image with lines drawn on it.
    Should be a blank image (all black) with lines drawn on it.
    
    `initial_img` should be the image before any processing.
    
    The result image is computed as follows:
    
    initial_img *  + img *  + 
    NOTE: initial_img and img must be the same shape!
    """
    return cv2.addWeighted(initial_img, , img, , )

def filter_white_yellow_colours(image):
    """
    Filter the image to include only yellow and white pixels to stop spurious
    line segments being found on the road.
    Idea and code example thanks to:
    https://solarianprogrammer.com/2015/05/08/detect-red-circles-image-using-opencv/
    and
    http://stackoverflow.com/questions/9179189/detect-yellow-color-in-opencv
    """
    # Filter white pixel range
    lower_white = np.array([200, 200, 200])
    upper_white = np.array([255, 255, 255])
    white_mask = cv2.inRange(image, lower_white, upper_white)
    white_image = cv2.bitwise_and(image, image, mask=white_mask)

    # Filter yellow pixel range
    hsv_image = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
    lower_yellow = np.array([20, 100, 100])  # this colour mask is bordering from gray to yellow
    upper_yellow = np.array([120, 255, 255])
    yellow_mask = cv2.inRange(hsv_image, lower_yellow, upper_yellow)
    yellow_image = cv2.bitwise_and(image, image, mask=yellow_mask)

    # Combine the two above images
    image2 = cv2.addWeighted(white_image, 1., yellow_image, 1., 0.)

    return image2

def create_masking_polygon(bounding_shape):
    '''
     Create masked edges using trapezoid-shaped region-of-interest
     e.g.
          ---
         /   \
        /     \
        -------
    '''
    #imshape = image.shape
    imshape = bounding_shape
    w = imshape[1]
    h = imshape[0]
    
    # Create vertices from bottom left trapezoid point clockwise...
    #vertices = np.array([[\
    #    ((w * (1 - mask_region_bottom_width)) // 2, h - (h*mask_region_bottom)), ((w * (1 - mask_region_top_width)) // 2, h - h * mask_region_height),\
    #    (w - (w * (1 - mask_region_top_width)) // 2, h - h * mask_region_height), (w - (w * (1 - mask_region_bottom_width)) // 2, h - (h*mask_region_bottom))]], dtype=np.int32)

    vertices = np.array([[\
        ((w * (1 - mask_region_bottom_width)) // 2, h),\
        ((w * (1 - mask_region_top_width)) // 2, h - h * mask_region_height),\
        (w - (w * (1 - mask_region_top_width)) // 2, h - h * mask_region_height),\
        (w - (w * (1 - mask_region_bottom_width)) // 2, h),
        (w - (w * (1 - mask_region_bottom_width)) // 2 - 50, h),
        (w - (w * (1 - mask_region_bottom_width)) // 2 - 50, h - 50),
        ((w * (1 - mask_region_bottom_width)) // 2 + 50, h - 50),
        ((w * (1 - mask_region_bottom_width)) // 2 + 50, h)]], dtype=np.int32)

    return vertices
```

```python
def hightlist_road_lines_in_image(source_image):
    """ Apply image processing pipeline to given image (Numpy array)
         - Colour filter to keep yellow and white only (helps with colour changes in road surface and shadows)
         - Grayscale
         - Smooth / Blur (like a lowpass filter to get rid of noise)
         - Canny edge detection
         - Select region of interest (trapezoid shape from the bottom of the image)
         - Hough Transform to detect straight lines
         - Combine images (lines from Hough and original image) as annotated output - a weighted
           combination is used so that the lines are partially transparent.
         
        Return the annotated image as a Numpy array """
    
    # Only keep white and yellow pixels, all others set to black - this provides
    # better accuracy on the challenge video where the yellow line goes haywire
    # due to patches in the road of differing colours. It is not required for the
    # other videos.
    #image = filter_white_yellow_colours(source_image)
    image = source_image
    
    # Read in and grayscale the image
    gray = grayscale(image)
    
    # Apply Gaussian smoothing
    blur = gaussian_blur(gray, kernel_size)
    
    # Apply Canny Edge Detector
    #edges = canny(blur, low_threshold, high_threshold)
    edges = auto_canny(blur)
    
    vertices = create_masking_polygon(image.shape)
        
    # Get our trapezoidal region of interest so as to ignore all other edges in the image
    # that are not near our road lines.
    masked_edges = region_of_interest(edges, vertices)
    
    # Run Hough on edge detected image
    line_image = hough_lines(masked_edges, , , threshold, min_line_length, max_line_gap)
    
    # Draw lane lines on the original image
    initial_image = source_image.astype('uint8')
    annotated_image = weighted_img(line_image, initial_image)
    
    return annotated_image
```
